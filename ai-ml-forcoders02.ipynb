# Computer vision example - MNIST Fashion Dataset


```python
# import tensorflow


import tensorflow as tf
```

```python
# "load" the (data set is already part of the tensorflow library!)
data = tf.keras.datasets.fashion_mnist

(training_images, training_labels), (test_images, test_labels) = data.load_data()

test_images[2]
```

```python
# "Normalize" all 28x28 pixel greysclae pictures ba dividibg by 255
# thus each pixel became a value between 0 and and 1 instedad aof 0 and 255
training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
test_images[2]
```

```python
# define neural network for training
model = tf.keras.models.Sequential (
     [ 
          tf.keras.layers.Flatten(
              input_shape= (28,28)
          ),
          tf.keras.layers.Dense(
              128, activation = tf.nn.relu
          ),
          tf.keras.layers.Dense(
              10, activation = tf.nn.softmax
          )   
         
      ]
)
```

```python
# specifiy loos function and optimizer function
model.compile(
    optimizer= 'adam',
    loss= 'sparse_categorical_crossentropy',
    metrics= ['accuracy']
)
```

```python
# trainig the model over 5 epochs
model.fit(
    training_images,
    training_labels,
    epochs = 50
)
```

```python
# evaluate the model
model.evaluate(
    test_images,
    test_labels
)
```

```python
# exploring model output
classifications = model.predict(test_images)
print(classifications[0])
```
